Image can be drawn as a histogram. Application of histogram => Analysis, thresholding, brightness and contrast, equalize an image.

Thresholding<br>
Partitioning an image into foreground and background. By making it black and white. We apply a simple law:<br>
portion pixel value > threshold = 255 (white)<br>
portion pixel value < threshold = 0 (black)<br>
Thresholding is the simplest method of image segmentation. Isolate objects (object detection, face detection etc)<br>
First convert to gray scale then convert.<br>

Thresholding categories:<br>
- Global or histogram based: Good for uniform backgrounds.
- Local or adaptive: For uneven background illumination.

To get the best threshold value, we can try skimage.filters.try_all_threshold.
